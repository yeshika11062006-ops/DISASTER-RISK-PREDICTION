```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Load the CSV
data_path = '../data/disaster_data.csv'
df = pd.read_csv(data_path)
df.head()
```

```python
import os
os.makedirs('../results', exist_ok=True)  # make sure folder exists

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='disaster_type', order=df['disaster_type'].value_counts().index)
plt.title('Count of Disasters by Type')
plt.xticks(rotation=45)

plt.savefig('../results/disaster_count.png', bbox_inches='tight')  # save before show
plt.show()
plt.close()
```
